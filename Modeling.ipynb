```python
import re
import nltk

# We need this dataset in order to use the tokenizer
from nltk.tokenize import word_tokenize

from nltk.corpus import stopwords

from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()

def process_text(text):
    # Make all the strings lowercase and remove non alphabetic characters
    text = re.sub('[^A-Za-z]', ' ', text.lower())

    # Tokenize the text; this is, separate every sentence into a list of words
    # Since the text is already split into sentences you don't have to call sent_tokenize
    tokenized_text = word_tokenize(text)

#     Remove the stopwords and stem each word to its root
    clean_text = [
        stemmer.stem(word) for word in tokenized_text
        if word not in stopwords.words('english')
    ]

    # Remember, this final output is a list of words
    return clean_text



sample = '''GUESS what after replace the rotors and pad CV SHAFT hub bear all 4 WHEEL cover and re grease the WHOLE 
            front suspension the loud annoying left front end noise be back i have 3 independent mechanic that have tell me 
            to let the dealership know they may want to replace the caliper as it sound like a faulty piston it only make the noise when 
            the car be in a lot of STOP and GO TRAFFIC and at SPEEDS of 5 15 mile PER hour it be a metal on METAL NOISE once the car be out 
            of the heavy braking traffic the noise goes away transitowne in WILLIAMSVILLE NY mechanic service ADVISER s and manger refuse to 
            listen to me as they say since there be no blue SPOTS on the rotors there be nothing wrong with the caliper for the LOVE of GOD you 
            have replaced everything else and it HASN t worked DUMB ass now the CAR be there for electrical PROBLEMS too and my daughter and 
            i almost got into a bad ACCIDENT because of this car CAN you say dodge figure out how to MAKE a HUGE lemon of course 
            i have a lawyer involve now because CHRYSLER do not want to cooperate with the lemon law in new york state STAY away from 
            CHRYSLER DODGE automobile BIGGEST HEADACHE EVER gsable'''

sample = sample.split()
texts = [process_text(text) for text in sample]
dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]
# texts
```

```python
from gensim import models
model = models.ldamodel.LdaModel(corpus, num_topics=3, id2word=dictionary, passes=15)


topics = model.print_topics(num_words=10)
for topic in topics:
    print(topic)
```
