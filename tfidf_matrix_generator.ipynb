```python
import pandas as pd

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation as LDiA


files = ['Audi.csv', 'BMW.csv', 'Chrysler.csv', 'Ford.csv', 'GMC.csv', 'Honda.csv', 
         'MB.csv', 'Nissan.csv', 'Toyota.csv']
```

```python
def topic_generator_tfidf(file):
    df = pd.read_csv('Make_data/{}'.format(file))
    df = df.dropna(axis=0, how='any')
    df.reset_index(inplace=True)

    corpus = []
    for i in range(df.shape[0]):
        corpus.append(df['Comments'][i])
    
#   BOW
    tfidf_vect = TfidfVectorizer()
    df_tfidf = pd.DataFrame(data=tfidf_vect.fit_transform(raw_documents=corpus).toarray())
    column_nums, terms = zip(*sorted(zip(tfidf_vect.vocabulary_.values(), tfidf_vect.vocabulary_.keys())))
    df_tfidf.column_nums = terms
    
#   ldia generate topics
    ldia = LDiA(n_components=10)
    ldia.fit_transform(df_tfidf)
    
    df_topics = pd.DataFrame(data=ldia.components_.T, index=terms)
#     return df_topics
        
    df_topics.to_csv('Make_data/topics/{}_tfidf_topics.csv'.format(file.split('.')[0]))
    
    
for file in files:
    topic_generator_tfidf(file)
```

```python
def tfidf_generator(file):
    df = pd.read_csv('Make_data/test/{}_test.csv'.format(file.split('.')[0]))
#     return df

    corpus = []
    for i in range(df.shape[0]):
        corpus.append(df['Comments'][i])
    
#   BOW
    tfidf_vect = TfidfVectorizer()
    df_tfidf = pd.DataFrame(data=tfidf_vect.fit_transform(raw_documents=corpus).toarray(), 
                          columns=tfidf_vect.get_feature_names(),)

        
    df_tfidf.to_csv('Make_data/test/{}_score.csv'.format(file.split('.')[0]), index=False)

for file in files:
     tfidf_generator(file)
```
